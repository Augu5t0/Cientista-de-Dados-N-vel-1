## Explorando as variáveis categóricas.

```python
def gerar_grafico_barras_pizza(dados, coluna):

  fig, ax = plt.subplots(1, 2, figsize=(15, 6))

  contagem = dados[coluna].value_counts()

  sns.barplot(x=contagem.index, y=contagem.values, ax=ax[0], palette='viridis', hue=contagem.index)
  ax[0].set_title(f'Distribuição de {coluna} (Barras)')
  ax[0].set_xlabel(coluna)
  ax[0].set_ylabel('Frequência')
  ax[0].tick_params(axis='x', rotation=45)

  ax[1].pie(contagem, labels=contagem.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis'))
  ax[1].set_title(f'Distribuição de {coluna} (Pizza)')

  plt.tight_layout()
  plt.show()
```

```python
gerar_grafico_barras_pizza(dados, 'Gender')
```

```python
gerar_grafico_barras_pizza(dados, 'Country')
```

```python
gerar_grafico_barras_pizza(dados, 'Sleep_Quality')
```

### A maioria dos participantes relata uma qualidade de sono entre "Boa" e "Regular". O gráfico de pizza permite visualizar rapidamente que os extremos ("Excelente" ou "Ruim") são menos frequentes, sugerindo que a maior parte da população mantém um padrão de sono mediano.

```python
gerar_grafico_barras_pizza(dados, 'Stress_Level')
```

### Assim como o sono, o estresse tende a se concentrar nos níveis Baixo e Médio. O gráfico de barras ajuda a identificar se existe algum grupo que se destaca, enquanto o de pizza mostra a proporção de pessoas sob alto estresse

```python
gerar_grafico_barras_pizza(dados, 'Occupation')
```

### A distribuição entre as ocupações parece uniforme

## Investigando possíveis correlações

```python
corr = dados.select_dtypes(include=['float64', 'int64']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlação entre Variáveis Numéricas')
plt.show()
```

### A maioria das cores no gráfico tende para tons neutros, o que confirma sua observação de que as variáveis são, em sua maioria, independentes ou possuem correlações muito fracas.

### A relação entre Caffeine_mg (ou Coffee_Intake) e Sleep_Hours é o ponto de maior interesse, apresentando aquele valor de -0.20. Isso indica que, à medida que a ingestão de cafeína aumenta, há uma leve tendência de queda nas horas de sono

```python
def gerar_grafico_correlacao(dados, colunaX, colunaY):
  plt.figure(figsize=(10, 6))
  sns.regplot(data=dados, x=colunaX, y=colunaY,
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
  plt.title(f'Correlação entre {colunaX} e {colunaY}')
  plt.xlabel(colunaX)
  plt.ylabel(colunaY)
  plt.show()
```

```python
gerar_grafico_correlacao(dados, 'Caffeine_mg', 'Sleep_Hours')
```

```python
gerar_grafico_correlacao(dados,'Caffeine_mg', 'Heart_Rate')
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

sns.barplot(data=resumo_stress, x='Nível de Estresse', y='Média de Cafeína (mg)', ax=ax1, palette='coolwarm',hue='Nível de Estresse', alpha=0.7)
ax1.set_ylabel('Média de Cafeína (mg)', color='darkred', fontsize=12)
ax1.tick_params(axis='y', labelcolor='darkred')

# Criando um segundo eixo Y para o Sono
ax2 = ax1.twinx()
sns.lineplot(data=resumo_stress, x='Nível de Estresse', y='Média de Sono (horas)', ax=ax2, marker='o', color='blue', linewidth=3)
ax2.set_ylabel('Média de Horas de Sono', color='blue', fontsize=12)
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Relação Estresse: Cafeína Ingerida vs. Horas de Sono', fontsize=14)
ax1.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# Idade e estresse em conjunto com consumo de café

df = dados.copy()

bins = [18, 30, 45, 60, 100]
labels = ['18-30', '31-45', '46-60', '60+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

consumo_cafe_idade_stress = df.groupby(['Age_Group', 'Stress_Level'])['Coffee_Intake'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(
    data=consumo_cafe_idade_stress,
    x='Age_Group',
    y='Coffee_Intake',
    hue='Stress_Level',
    palette='magma'
)

plt.title('Consumo Médio de Café por Faixa Etária e Nível de Estresse')
plt.xlabel('Faixa Etária')
plt.ylabel('Média de Xícaras de Café / Dia')
plt.legend(title='Nível de Estresse')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Principais Descobertas

### **Relação entre Cafeína, Estresse e Sono**
Os dados mostram uma correlação clara entre o aumento do estresse, o consumo de cafeína e a privação de sono:

* **Impacto no Sono:** Clientes com alto nível de estresse (que também consomem as maiores doses de cafeína) dormem, em média, 2,8 horas a menos do que aqueles com baixo nível de estresse.

    * Média de sono no Estresse Baixo: 7,25 horas.

     * Média de sono no Estresse Alto: 4,45 horas.

* **Consumo de Cafeína:** O grupo sob alto estresse consome, em média, 283 mg de cafeína por dia, enquanto o grupo de baixo estresse consome cerca de 225 mg.

### **O Nível de Estresse e a Qualidade do Sono**
A análise de matriz de frequência (crosstab) revelou que o nível de estresse é um determinante crítico da qualidade percebida do sono:

* **Estresse Alto:** 100% dos indivíduos nesta categoria relataram qualidade de sono **"Poor"** (Ruim).

* **Estresse Médio:** Está totalmente associado à qualidade de sono **"Fair"** (Regular).

* **Estresse Baixo:** É o único grupo que consegue atingir as categorias **"Good" **(Boa) e "Excellent" (Excelente).

### **Outras Descobertas Relevantes**
* **Gênero:** Não há diferença significativa no tempo de sono entre homens e mulheres (ambos com média de 6,64 horas).

* **Frequência Cardíaca**: Embora se espere que o estresse eleve os batimentos, a diferença média foi pequena: 70,9 bpm no estresse alto vs. 70,4 bpm no estresse baixo.

* **Idade e Qualidade de Sono:** A idade média dos participantes é praticamente a mesma em todas as categorias de qualidade de sono (em torno de 35 anos), indicando que, neste dataset, envelhecer não significa necessariamente dormir pior.
