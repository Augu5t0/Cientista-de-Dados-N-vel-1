## Explorando as variáveis categóricas.

```python
def gerar_grafico_barras_pizza(dados, coluna):

  fig, ax = plt.subplots(1, 2, figsize=(15, 6))

  contagem = dados[coluna].value_counts()

  sns.barplot(x=contagem.index, y=contagem.values, ax=ax[0], palette='viridis', hue=contagem.index)
  ax[0].set_title(f'Distribuição de {coluna} (Barras)')
  ax[0].set_xlabel(coluna)
  ax[0].set_ylabel('Frequência')
  ax[0].tick_params(axis='x', rotation=45)

  ax[1].pie(contagem, labels=contagem.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis'))
  ax[1].set_title(f'Distribuição de {coluna} (Pizza)')

  plt.tight_layout()
  plt.show()
```

```python
def gerar_grafico_correlacao(dados, colunaX, colunaY):
  plt.figure(figsize=(10, 6))
  sns.regplot(data=dados, x=colunaX, y=colunaY,
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
  plt.title(f'Correlação entre {colunaX} e {colunaY}')
  plt.xlabel(colunaX)
  plt.ylabel(colunaY)
  plt.show()
```

#Modelo preditivo

## pré-processamento dos dados

```python
df_copia = dados.copy()
```

```python
df_copia['Caffeine_per_Hour'] = df_copia['Caffeine_mg'] / (df_copia['Sleep_Hours'])
```

```python
mapping = {'Poor': 0, 'Fair': 1, 'Good': 2, 'Excellent': 3}
df_copia['Sleep_Quality'] = df_copia['Sleep_Quality'].map(mapping)
```

```python
df_copia.head()
```

```python
# removendo colunas desnecessárias e definindo a váriavel alvo
X = df_copia.drop(columns = ['ID', 'Coffee_Intake', 'Gender','Country', 'Sleep_Quality' ])
y = df_copia['Sleep_Quality']
```

```python
y.value_counts(normalize=True)*100 # Conferindo se há um desbalanceamento das categorias na variável alvo
```

### **Por que remover essas colunas ?**

**ID:** Esta coluna é apenas um identificador único para cada linha. Ela não carrega nenhuma informação estatística ou padrão que ajude o modelo a prever o comportamento de novos dados. Mantê-la seria apenas adicionar "ruído" ao modelo.

**Coffee_Intake:** Note que você tem duas colunas que medem basicamente a mesma coisa: Coffee_Intake e Caffeine_mg. Ter as duas gera multicolinearidade (informação redundante). Como a coluna Caffeine_mg é numérica contínua e mais precisa

**Gender:**  "não há diferença significativa no tempo de sono entre homens e mulheres" (média de 6,64 horas para ambos). O objetivo principal é prever a qualidade do sono e os dados mostram que o gênero não influencia esse resultado, removê-lo simplifica o modelo.

**Country:** como existem 20 países diferentes, manter essa coluna exigiria a criação de muitas variáveis dummy (One-Hot Encoding), o que aumentaria a complexidade do modelo sem necessariamente aumentar a precisão.

```python
X
```

```python
# Aplicando OneHotEnconder nas colunas categóricas
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

colunas_transformadas = ColumnTransformer(transformers=[
    ('encoder', OneHotEncoder(drop='first', sparse_output=False),
     ['Stress_Level', 'Health_Issues', 'Occupation'])
], remainder='passthrough')

dados_processados = colunas_transformadas.fit_transform(X)

X = pd.DataFrame(dados_processados, columns=colunas_transformadas.get_feature_names_out())
X
```

```python
from sklearn.preprocessing import StandardScaler

colunas_para_escalar = [
    'remainder__Age',
    'remainder__Caffeine_mg',
    'remainder__Sleep_Hours',
    'remainder__BMI',
    'remainder__Heart_Rate',
    'remainder__Physical_Activity_Hours',
    'remainder__Caffeine_per_Hour'
]

scaler = StandardScaler()
X[colunas_para_escalar] = scaler.fit_transform(X[colunas_para_escalar])
X.head()
```

## Divisão dos dados em treino e teste

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42) # stratify=y para garantir a proporção das classes no conjuto de treino e teste
```

## Treinando modelos

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_lr = LogisticRegression(multi_class='multinomial', class_weight='balanced', max_iter=1000)
modelo_lr.fit(X_treino, y_treino)
y_pred_lr = modelo_lr.predict(X_teste)

print(f"Acurácia Treino: {modelo_lr.score(X_treino, y_treino):.2f}")
print(f"Acurácia Teste: {modelo_lr.score(X_teste, y_teste):.2f}")
```

```python
modelo_rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
modelo_rf.fit(X_treino, y_treino)
y_pred_rf = modelo_rf.predict(X_teste)

print(f"Acurácia Treino: {modelo_rf.score(X_treino, y_treino):.2f}")
print(f"Acurácia Teste: {modelo_rf.score(X_teste, y_teste):.2f}")
```

## Avaliando modelos

### Matriz de confusão

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# matriz de confusão do dos resultados do modelo de regressão logística
matriz_confusao = confusion_matrix(y_teste,y_pred_lr)

visualizacao = ConfusionMatrixDisplay(confusion_matrix=matriz_confusao, display_labels=modelo_lr.classes_)
visualizacao.plot();
```

```python
ConfusionMatrixDisplay.from_predictions(y_teste, y_pred_lr, normalize='true', cmap='Blues');
```

```python
# matriz de confusão do dos resultados do modelo de regressão logística
matriz_confusao = confusion_matrix(y_teste,y_pred_rf)

visualizacao = ConfusionMatrixDisplay(confusion_matrix=matriz_confusao, display_labels=modelo_rf.classes_)
visualizacao.plot();
```

```python
ConfusionMatrixDisplay.from_predictions(y_teste, y_pred_rf, normalize='true', cmap='Blues');
```

```python
# Variavesi mais importantes para o modelo - Random Forest
importancias = pd.Series(modelo_rf.feature_importances_, index=X_treino.columns)
importancias.sort_values(ascending=False).head(10).plot(kind='barh')
plt.title("Top 10 Variáveis Mais Importantes")
plt.show()
```

```python
nomes_colunas = colunas_transformadas.get_feature_names_out()

df_coeficientes = pd.DataFrame(
    modelo_lr.coef_,
    columns=nomes_colunas,
    index=[f'Classe {i}' for i in range(len(modelo_lr.classes_))]
)
display(df_coeficientes.T)
```

* **remainder__Sleep_Hours:** Tem o coeficiente mais alto para a Classe 3 (10.97). Isso confirma que as horas de sono são o maior preditor de uma qualidade de sono "Excellent".

* **encoder__Stress_Level_Low:** Tem um peso forte (4.95) para a Classe 2, indicando que baixo estresse é importante para um sono "Good".

### Relátorio de métricas

## Relatório da avaliação

### **Qual modelo performou melhor?**

* **Random Forest:** Alcançou 100% de acurácia no treino e 99% no teste. Ele foi capaz de capturar todas as nuances dos dados, mas com uma complexidade maior.

* **Regressão Logística:** Alcançou 99% de acurácia tanto no treino quanto no teste.

* **Veredito:** Embora a Random Forest tenha se saido 'melhor', a Regressão Logística é o melhor modelo para este caso por ser um modelo mais simples que entrega o mesmo resultado de um modelo complexo.

### **Há overfitting ou underfitting?**

* **Random Forest:** Apresenta um leve sinal de overfitting, evidenciado pelos 100% no treino contra 99% no teste. Isso indica que o modelo "decorou" perfeitamente o conjunto de treino. No entanto, como o resultado no teste ainda é altíssimo, o modelo generaliza muito bem.

* **Regressão Logística:** Apresenta um ajuste ideal. Os resultados de treino e teste serem praticamente idênticos (99%).

* **Underfitting:** Não foi detectado em nenhum dos modelos, dado que ambos conseguiram aprender os padrões complexos do dataset.

### **Por que os modelos performaram tão bem?**

* **Qualidade e Separação dos Dados**: O dataset apresenta uma separação muito clara entre as categorias de qualidade de sono. Variáveis como Sleep_Hours (Horas de Sono) e Stress_Level (Nível de Estresse) possuem uma correlação direta e quase matemática com o alvo. Conforme vimos nos coeficientes, um aumento nas horas de sono é um preditor quase infalível de um sono "Excellent".
